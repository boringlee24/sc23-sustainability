```python
random.seed(1234)
##plotting styles
fig = mp.figure(figsize=(10,2.7))
mp.subplots_adjust(wspace=0.3, hspace=0.2)
plt.rcParams.update({'mathtext.default':  'regular' })
from matplotlib.ticker import MultipleLocator

manufacturing=[27502.69029,22456.568,16049.21643,15577.54971,14231.34171,9168.731429]
packaging=[3900,3900,3900,1350,1350,150]
tot=[manufacturing[i] + packaging[i] for i in range(len(manufacturing))]
x=["AMD\nMI250X","NVIDIA\nA100","NVIDIA\nV100","AMD EPYC\n7763","AMD EPYC\n7742","Intel Xeon\nGold 6240R"]
##plotting
barWidth=0.65
ax = mp.subplot(121)
ax.grid(axis='y', ls='dotted', zorder=1000)
tot = np.array(tot) / 1000
ax.bar([i for i in range(len(tot))], tot , width = barWidth, color='#fdb42f',edgecolor = 'black',capsize=15, label='Manufacturing', zorder=3)
ax.tick_params(labelsize=15)
ax.set_ylim([0,40])
ax.yaxis.set_major_locator(MultipleLocator(10))
# ax.set_yticks([0,7000,14000,21000,28000,35000])
ax.set_xticks([i for i in range(len(tot))])
ax.set_xticklabels(x, fontsize=14, rotation=90)
ax.set_ylabel("Embodied\nCarbon (kg $CO_{2}$)", fontsize=14)
ax.axvline(x=2.5, linewidth=1, linestyle='--', color='black')
ax.axvline(x=5.5, linewidth=1, linestyle='--', color='black')
ax.text(-0.4,35,"GPUs", fontsize=14,color='black')
ax.text(2.65,35,"CPUs", fontsize=14,color='black')
#ax.text(6,29000,"Memory/HDD", fontsize=15,color='black')
ax.set_xlim([-0.5,5.5])


tot= np.array(tot) / np.array([47.9, 9.7, 7.8, 2.5, 2.3, 0.5625])
# x=["A100 PCIe\n40GB","AMD INSTINCT\nMI250X","V100 SXM2\n32GB","AMD EPYC\n3rd Gen. 64C","AMD EPYC\n7742","Intel Xeon(R)\nGold 6240R"]
##plotting
barWidth=0.65
ax1 = mp.subplot(122)
ax1.grid(axis='y', ls='dotted', zorder=1000)
ax1.bar([i for i in range(len(tot))], tot , width = barWidth, color='#0d0887',edgecolor = 'black',capsize=15, label='Manufacturing', zorder=3)
ax1.tick_params(labelsize=15)
ax1.set_ylim([0,18])
ax1.yaxis.set_major_locator(MultipleLocator(4))
# ax1.set_yticks([0,10,20,30,40,50])
ax1.set_xticks([i for i in range(len(tot))])
ax1.set_xticklabels(x, fontsize=14, rotation=90)
ax1.set_ylabel("Embodied Carbon\n(kg $CO_{2}$) per TeraFLOPS", fontsize=14, y=0.45)
ax1.axvline(x=2.5, linewidth=1, linestyle='--', color='black')
ax1.text(-0.4,15.5,"GPUs", fontsize=14,color='black')
ax1.text(2.65,15.5,"CPUs", fontsize=14,color='black')
#ax1.text(6,29000,"Memory/HDD", fontsize=15,color='black')
ax1.set_xlim([-0.5,5.5])

plt.savefig("../figures/cpugpu_embidied_carbon.pdf", bbox_inches="tight")
```
